```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from PIL import Image
```

```python
path = os.path.join('..', 'data', 'ImageNetSubset', 'train')
print('Training path:', path, '\n')

class_names = os.listdir(path)
print('Class folder names:', class_names, '\n')

for c in class_names:
    path =  os.path.join('..', 'data', 'ImageNetSubset', 'train', c)
    images = os.listdir(path)

    for i in images[:10]: # only first 3 each
        path = os.path.join('..', 'data', 'ImageNetSubset', 'train', c, i)
        print(f'Image ({c}) path: {path}')

        with Image.open(path) as image:
            print('Resolution:', image.size, '\n')
    
```
